# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging
- Add more tools


````python
import json
import logging
````

````python
logging.basicConfig(level=logging.INFO, format="%(asctime)s | %(levelname)s | %(message)s")
logger = logging.getLogger("SmartAgent")
````

````python
# 🛠️ TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
def agent(query: str):
    if not isinstance(query, str) or not query.strip():
        logger.warning("Validation failed: empty or non-string query")
        return {"type": "error", "result": "Invalid or empty query received"}

    query_lower = query.lower()
    logger.info(f"Validating and routing query: {query}")

    try:
        if "calculate" in query_lower:
            expression = query_lower.split("calculate", 1)[1].strip()
            if not expression:
                logger.warning("No expression found after 'calculate'")
                return {"type": "error", "result": "No expression found to calculate"}

            result = calculator(expression)
            if result == "Error in calculation":
                logger.error(f"Calculator failed to parse expression: '{expression}'")
                return {"type": "error", "result": "Error in calculation"}

            logger.info(f"Calculator tool succeeded: {expression} -> {result}")
            return {"type": "calculation", "result": result}

        elif "keywords" in query_lower:
            text = query
            for phrase in ["extract keywords from", "keywords from", "extract keywords", "keywords"]:
                if phrase in text.lower():
                    idx = text.lower().find(phrase)
                    text = text[:idx] + text[idx + len(phrase):]
            keywords = extract_keywords(text.strip())
            logger.info(f"Keyword tool succeeded: {keywords}")
            return {"type": "keywords", "result": keywords}

        else:
            logger.info("No tool match found, routing to general fallback")
            return {"type": "general", "result": f"You asked: '{query}'. This is a general query handled directly."}

    except Exception as e:
        logger.error(f"Unexpected pipeline failure: {e}")
        return {"type": "error", "result": f"Something went wrong: {str(e)}"}
````

## 📦 Expected Output Format

```
{
  "type": "calculation / keywords / general / error",
  "result": ...
}
```

````python
# 🧪 Test Cases

queries = [
    "Calculate 20 + 5",
    "Extract keywords from Artificial Intelligence is transforming industries",
    "What is machine learning?"
]

for q in queries:
    print("Query:", q)
    print("Response:", agent(q))
    print("-" * 50)
````

````python
# 🎯 Interactive Mode

while True:
    user_input = input("Enter query (type 'exit' to stop): ")
    if user_input.lower() == "exit":
        break
    print("Response:", agent(user_input))
````
